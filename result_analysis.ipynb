```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
metrics = ['MAE', 'RMSE', 'Precision', 'Recall', 'F-measure']
svd_metrics = [0.6659244588258377, 0.8704853648298398, 0.9253354178847643, 0.9657490874422522, 0.9451104198494533]
nmf_metrics = [0.7101148582937473, 0.9278849573626533, 0.922696573537139, 0.9444908277670366, 0.9334665067996909]
co_clustering_metrics = [0.7308592818509391, 0.9450504684603249, 0.923974879756064, 0.9531359833883478, 0.938328921252078]
```

```python
x = np.arange(len(metrics))
width = 0.25
fig, ax = plt.subplots()
ax.bar(x - width, svd_metrics, width, label='SVD')
ax.bar(x, nmf_metrics, width, label='NMF')
ax.bar(x + width, co_clustering_metrics, width, label='CoClustering')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()
ax.set_ylabel('Error')
ax.set_title('Comparison of Algorithms')
fig.set_size_inches(9, 6)
plt.show()
```
