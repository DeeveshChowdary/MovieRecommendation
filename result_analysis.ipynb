```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
metrics = ['MAE', 'RMSE', 'Precision', 'Recall', 'F-measure']
svd_metrics = [0.6659244588258377, 0.8704853648298398, 0.7165573770491803, 0.7165573770491803, 0.7165573770491803]
nmf_metrics = [0.7101148582937473, 0.9278849573626533, 0.7008196721311475, 0.7008196721311475, 0.7008196721311475]
co_clustering_metrics = [0.7308592818509391, 0.9450504684603249, 0.7049180327868853, 0.7049180327868853, 0.7049180327868853]
```

```python
x = np.arange(len(metrics))
width = 0.25
fig, ax = plt.subplots()
ax.bar(x - width, svd_metrics, width, label='SVD')
ax.bar(x, nmf_metrics, width, label='NMF')
ax.bar(x + width, co_clustering_metrics, width, label='CoClustering')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()
ax.set_ylabel('Error')
ax.set_title('Comparison of Algorithms')
plt.show()
```
